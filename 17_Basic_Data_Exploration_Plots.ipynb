```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(0, 1, 1000)  # Generate random data
plt.hist(data, bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.show()
```
