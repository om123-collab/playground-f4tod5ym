```python
import matplotlib.pyplot as plt

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def insert(root, data):
    # ... (Implementation of BST insertion) ...
    # This is a placeholder for the actual insertion logic.
    # You'll need to replace this comment with the code.
    if root is None:
        return Node(data)
    if data < root.data:
        root.left = insert(root.left, data)
    else:
        root.right = insert(root.right, data)
    return root

def visualize_bst(root, ax):
    if root:
        ax.text(root.data, 0, str(root.data))
        visualize_bst(root.left, ax)
        visualize_bst(root.right, ax)

# ... (Example: Create a simple BST) ...
root = Node(5)
insert(root, 3)
insert(root, 7)

fig, ax = plt.subplots()
visualize_bst(root, ax)
plt.show()
```
